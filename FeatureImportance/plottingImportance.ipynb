```python
import pickle
```

```python
feature_set='hidden_states'
```

```python
model = pickle.load(open(f'/content/drive/MyDrive/CancerVsLaryngitis/CreateModels/FeatureImportance/{feature_set}.sav', 'rb'))
```

```python
import matplotlib.pyplot as plt
```

```python
sum_list = [sum(x) for x in zip(model['Accuracy'], model['Recall'], model['Precision'])]
average_list = [x / 3 for x in sum_list]
```

```python
len(average_list)
```

```python
average_list.index(max(average_list))
```

```python
model['Accuracy'][average_list.index(max(average_list))]
```

```python
model['Precision'][average_list.index(max(average_list))]
```

```python
model['Recall'][average_list.index(max(average_list))]
```

```python
plt.plot(model['Accuracy'][:-1], label='Accuracy')
plt.plot(model['Recall'][:-1], label='Recall')
plt.plot(model['Precision'][:-1], label='Precision')
plt.axvline(x = average_list.index(max(average_list)), color='r', label = 'Best feature set')

plt.xlabel('Number of features used to train the model')
plt.legend()

plt.savefig(f'/content/drive/MyDrive/CancerVsLaryngitis/CreateModels/FeatureImportance/{feature_set}.png')
```
