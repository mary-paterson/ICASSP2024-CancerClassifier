```python
import librosa
import os
import numpy as np
import pandas as pd
```

```python
columns = list(range(20))
columns.append('file')
columns.append('pathology')
print(columns)
```

```python
signals = []
df = pd.DataFrame(columns=columns)

counter = 0

dir = '/content/drive/MyDrive/CancerVsLaryngitis/TestSet/'

for subdir, dirs, files in os.walk(dir):
  for file in files:
    if file.endswith('.wav'):
      counter+=1
      y, sr = librosa.load(subdir+'/'+file, sr=16000)
      signals.append(y)
      mfcc = librosa.feature.mfcc(y=y, sr=sr)
      mfcc_feat = np.mean(mfcc, axis=1)
      pathology = subdir.split('/')[-1]

      row = list(mfcc_feat).copy()
      row.append(file)
      row.append(pathology)

      df.loc[len(df)]=row
```
